```python
import pandas as pd
```

```python
joints_range_angle = {"lumbar_extension": (-15,90),
                          "lumbar_bending": (-15,15),
                          "lumbar_rotation": (-90,90),
                          "arm_flex_r": (-30,180),
                          "arm_add_r": (-90,90),
                          "arm_rot_r": (-90,90),
                          "elbow_flex_r": (0,90),
                          "pro_sup_r": (-30,30),
                          "arm_flex_l": (-30,180),
                          "arm_add_l": (-90,90),
                          "arm_rot_l": (-90,90),
                          "elbow_flex_l": (0,90),
                          "pro_sup_l": (-90,90),
                          }
for key,value in enumerate(joints_range_angle.items()):
    joint = value[0]
    range = value[1][1]-value[1][0]
    print(joint,range)
```

```python
data_window = pd.read_csv("../../../final-dataset/S01-A05-T01-0Â°-1.csv")
data_window.head()
```

```python
column_names = data_window.columns.values
for column in column_names:
    min = joints_range_angle[column][0]
    max = joints_range_angle[column][1]
    data_window[column] = data_window[column]/(max-min)
```
